```python
data = pd.read_csv('prod_prod.csv')
print(data.shape)
data_text =  data[['id_Familia', 'des_Articulo']] #desde BQ me asegure que cada Familia tenga descripcion 
print(data_text.shape)
data_text = data_text.dropna().reset_index()
```

```python
data_text = data_text.groupby(['id_Familia'])['des_Articulo'].apply(lambda x: ' | '.join(x)).reset_index()
data_text.columns = ['id_Familia', 'des_Familia']
data_text.des_Familia = data_text.des_Familia.apply(lambda x: x.lower())
data_text
```

```python
toktok = ToktokTokenizer()
stemmer = SnowballStemmer("spanish")
stops_spa = stopwords.words('spanish')
```

```python
def limpia ( indice):
    #tokens_clean = [ stemmer.stem(x) for x in toktok.tokenize(  data_text.des_Familia[ indice ] ) if x.isalnum() and x not in stops_spa ]
    tokens_clean = [ x for x in toktok.tokenize(  data_text.des_Familia[ indice ] ) if x.isalnum() and x not in stops_spa ]
    
    return(tokens_clean)
    
```

```python
corpus = list(map( lambda x: limpia(x), data_text.index ) ) 
texto = reduce(lambda x, y: x+y, corpus)
bolsa = Counter(texto)
```

```python
wc = WordCloud(background_color="white", max_words=100)
wc.generate_from_frequencies(bolsa)
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
demo = pd.DataFrame(data=bolsa.values(), columns=['y'])
plt.hist( np.log( demo.y ) )
print(demo.describe())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
vectorizer = CountVectorizer(stop_words=stops_spa)
y = data_text.id_Familia
#X_train, X_test, y_train, y_test = train_test_split( data_text['des_Familia'], y, test_size=0.0, random_state=0)
X_test = X_train = data_text['des_Familia']
y_train = y_test = y
```

```python
bolsa_train = vectorizer.fit_transform(X_train)
bolsa_test = vectorizer.transform(X_test)
bolsa_df = pd.DataFrame(bolsa_train.A, columns=vectorizer.get_feature_names())
```

```python
from sklearn import metrics
from sklearn.naive_bayes import MultinomialNB
bolsa_nb = MultinomialNB()
bolsa_nb.fit( bolsa_train , y_train)
pred = bolsa_nb.predict( bolsa_test)
score = metrics.accuracy_score(y_test, pred)
print(score)
cm = metrics.confusion_matrix(y_test, pred)
print(cm)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words=stops_spa, max_df=1.0, min_df=0.008)
tfidf_train = tfidf_vectorizer.fit_transform(X_train)
tfidf_test = tfidf_vectorizer.transform(X_test)
print(tfidf_train.shape)
tfidf_df = pd.DataFrame(tfidf_train.A, columns=tfidf_vectorizer.get_feature_names())
tf_nb= MultinomialNB()
tf_nb.fit(tfidf_train, y_train)
pred = tf_nb.predict( tfidf_test )
score = metrics.accuracy_score(y_test, pred)
print(score)
cm = metrics.confusion_matrix(y_test, pred )
print(cm)
```

```python
Probs  = tf_nb.predict_proba( tfidf_test )
```

```python
Probs.shape
```

```python
t = pd.DataFrame(data=Probs.flatten(), columns=['p'])
plt.hist(t.p )
t.describe()
```

```python
threshold = np.quantile(t.p, 0.98)
# 80 % => 687,122 aristas finales 
# 90 % => 295,844 aristas finales
# 95 % => 139,193 aristas finales
# 96 % => 110,602 aristas finales 
# 97 % => 82,117 aristas finales 
# 98 % => 54,099
# 99 % => 26,597 aristas finales 
threshold
```

```python
mask = Probs < threshold
mask
Probs[mask ] = 0
```
